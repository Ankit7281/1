```python
import face_recognition
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the known and test images
known_image = face_recognition.load_image_file("ankitp001.jpeg")
known_encoding = face_recognition.face_encodings(known_image)[0]

test_image = face_recognition.load_image_file("test.jpg")
test_locations = face_recognition.face_locations(test_image)
test_encodings = face_recognition.face_encodings(test_image, test_locations)

# Convert test image to OpenCV format (BGR)
image = cv2.cvtColor(test_image, cv2.COLOR_RGB2BGR)

# Compare each face in test image to the known face
for (top, right, bottom, left), face_encoding in zip(test_locations, test_encodings):
    match = face_recognition.compare_faces([known_encoding], face_encoding)[0]
    face_dist = face_recognition.face_distance([known_encoding], face_encoding)[0]

    # Draw box and label
    color = (0, 255, 0) if match else (0, 0, 255)
    label = "Match" if match else "No Match"
    cv2.rectangle(image, (left, top), (right, bottom), color, 2)
    cv2.putText(image, f"{label} ({face_dist:.2f})", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

# Convert BGR back to RGB for display in Colab
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show the result
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("Face Recognition Result")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Load image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load OpenCV's pre-trained Haar Cascade face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Draw boxes
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Convert to RGB and show
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Detected Faces')
plt.show()
```
