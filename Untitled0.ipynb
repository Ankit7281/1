```python
dataset_yaml = """
train: /content/drive/MyDrive/Dataset/train/images
val: /content/drive/MyDrive/Dataset/valid/images

nc: 2
names: ['MotorBike', 'NonMotorBike']

"""

with open('/content/dataset.yaml', 'w') as file:
    file.write(dataset_yaml)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model
#results = model.train(data="coco8.yaml", epochs=100, imgsz=640)
model.train(data='/content/dataset.yaml', epochs=50, batch=16, imgsz=640)
```

```python
# Save the trained model to your Google Drive
model.save('/content/drive/MyDrive/Dataset/helmet_detection_model.pt')
```

```python
# Evaluate the model on the validation set
metrics = model.val()
print(metrics)
```

```python
import os

# Folder containing images
folder_path = '/content/drive/MyDrive/TestImages'

# Get all image file paths from the folder (filtering common image types)
image_paths = [os.path.join(folder_path, fname)
               for fname in os.listdir(folder_path)
               if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Run inference on all images
results = model(image_paths)

# Show results for each image
for result in results:
    result.show()
```
